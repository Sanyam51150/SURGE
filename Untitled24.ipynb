```python
import numpy as np
import matplotlib.pyplot as plt

class MABandit:
    def __init__(self, num_bandits):
        self.num_bandits = num_bandits
        self.means = np.random.normal(0, 10, self.num_bandits)

    def pull_bandit(self, bandit):
        return np.random.normal(self.means[bandit], 10)

class MAML:
    def __init__(self, num_bandits):
        self.num_bandits = num_bandits
        self.alpha = 0.1
        self.beta = 1.0
        self.theta = np.zeros(self.num_bandits)

    def update_theta(self, rewards, selected_bandit):
        probabilities = np.exp(self.theta) / np.sum(np.exp(self.theta))
        gradient = np.zeros(self.num_bandits)
        gradient -= probabilities * rewards
        gradient[selected_bandit] = (1 - probabilities[selected_bandit]) * rewards+probabilities[selected_bandit] * rewards
        self.theta += self.alpha * gradient

    def choose_bandit(self):
        probabilities = np.exp(self.theta) / np.sum(np.exp(self.theta))
        return np.random.choice(self.num_bandits, p=probabilities)

class UCB:
    def __init__(self, num_bandits):
        self.num_bandits = num_bandits
        self.counts = np.zeros(self.num_bandits)
        self.values = np.zeros(self.num_bandits)

    def choose_bandit(self, step):
        exploration = np.sqrt(2 * np.log(step + 1) / (self.counts + 1e-6))
        ucb_values = self.values + exploration
        return np.argmax(ucb_values)

def main():
    num_bandits = 5
    num_steps = 100
    num_runs = 100

    pg_rewards = np.zeros((num_runs, num_steps))
    ucb_rewards = np.zeros((num_runs, num_steps))

    for run in range(num_runs):
        bandit = MABandit(num_bandits)
        pg_agent = MAML(num_bandits)
        ucb_agent = UCB(num_bandits)

        for step in range(num_steps):
            pg_bandit = pg_agent.choose_bandit()
            ucb_bandit = ucb_agent.choose_bandit(step + 1)

            pg_reward = bandit.pull_bandit(pg_bandit)
            ucb_reward = bandit.pull_bandit(ucb_bandit)

            pg_rewards[run, step] = pg_reward
            ucb_rewards[run, step] = ucb_reward

            pg_agent.update_theta(pg_reward, pg_bandit)
            ucb_agent.counts[ucb_bandit] += 1
            ucb_agent.values[ucb_bandit] += (ucb_reward - ucb_agent.values[ucb_bandit]) / ucb_agent.counts[ucb_bandit]

    pg_avg_rewards = np.mean(pg_rewards, axis=0)
    ucb_avg_rewards = np.mean(ucb_rewards, axis=0)

    plt.plot(range(num_steps), pg_avg_rewards, label='MAML')
    plt.plot(range(num_steps), ucb_avg_rewards, label='UCB')
    plt.xlabel('Steps')
    plt.ylabel('Average Reward')
    plt.legend()
    plt.show()

main()
```
